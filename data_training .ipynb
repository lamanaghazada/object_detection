```python
# Data Augementation techniques
import albumentations as A
# Define custom Albumentations transforms
custom_transforms = [
    A.Blur(blur_limit=7, p=0.5),
    A.GaussNoise(std_range=(0.04, 0.2), p=0.3),
    A.RandomBrightnessContrast(brightness_limit=0.2, contrast_limit=0.2, p=0.5),
    A.HueSaturationValue(hue_shift_limit=20, sat_shift_limit=30, val_shift_limit=20, p=0.5),
]
```

```python
# 1st model

model = YOLO('yolov8n.pt')
results_1 = model.train(
data= "/content/drive/MyDrive/dataset/data.yaml",
epochs = 50, # 50–100 aralığı
imgsz = 640,
batch = 16,
seed = SEED,
freeze=15,  # freeze first 10-15 layers  # without freezing got poorer results(f1,precision,recall)
patience=10,
lr0=0.0001, #first checked default lr(0.01) => got poor performance
optimizer='AdamW',
project = 'voc_detection',
name = '1st_',

# Albumentations
augmentations=custom_transforms,

# YOLO augmentations
    fliplr=0.5,
    scale=0.5,

)
```

```python
# 2nd model
model = YOLO('yolov8n.pt')
results = model.train(
data= "/content/drive/MyDrive/dataset/data.yaml",
epochs = 75, # 50–100
imgsz = 640,
batch = 16,
seed = SEED,
freeze=15,  # freeze first 10-15 layers  # without freezing got poorer results(f1,precision,recall)
patience=10,      #no big change when increment or decrement
lr0=0.0001,  #first default lr(0.01) =>poor performance
optimizer='AdamW',
project = 'voc_detection_task',
name = '2nd_try',

# Albumentations
augmentations=custom_transforms,

# YOLO augmentations
    fliplr=0.5,
    scale=0.5,
)
```
